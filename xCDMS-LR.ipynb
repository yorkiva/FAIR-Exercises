In this notebook we will use the superCDMS dataset introduced in the previous notebook and perform some simple machine learning exercises to address the regression problem. In particular, we want to explore the regression problem of predicting the impact location based on the timing measurements provided in our dataset.

Let's load our dataset.

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import *
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression, Ridge, Lasso
import seaborn as sns
import pandas as pd
from RegressionUtils import *
```

```python
## Get the dataset as numpy array

data_ext = pd.read_csv('data/CDMSData_extended.csv', sep=',')
data_red = pd.read_csv('data/CDMS_Dataset.csv', sep=',')
```

```python
Datasets = {'extended_noamp':   DataProcessor(data_ext, data_type = 'extended',
                                              preprocess = True, skip_amp = True),
            'extended_withamp': DataProcessor(data_ext, data_type = 'extended',
                                              preprocess = True, skip_amp = False),
            'simple':           DataProcessor(data_red, data_type = 'simple',
                                              preprocess = True, skip_amp = True)
           }
tagdict = {'extended_noamp': 'X-nA',
           'extended_withamp': 'X-A',
           'simple': 'S'
           }
```

```python
data_modes = ['simple', 'extended_noamp', 'extended_withamp']
reg_modes =  ['Full', 'PCA']
ridge_alphas = [1.e-6, 0.75, 1.33]
lasso_alphas = [1.e-6, 3e-5, 3e-5]

test_losses_LR = {}
test_losses_ridge = {}
test_losses_lasso = {}
val_losses_LR = {}
val_losses_ridge = {}
val_losses_lasso = {}
alpha_ridge = {}
alpha_lasso = {}
err_tags = []
terr_data = []
verr_data = []
alpha_tags = []
alpha_data = []

for mode in data_modes:
    dataset = Datasets[mode]
    for reg in reg_modes: 
        mode_tag = tagdict[mode] + '-' + reg
        test_losses_LR[mode_tag] = []
        test_losses_ridge[mode_tag] = []
        test_losses_lasso[mode_tag] = []
        val_losses_LR[mode_tag] = []
        val_losses_ridge[mode_tag] = []
        val_losses_lasso[mode_tag] = []
        alpha_ridge[mode_tag] = []
        alpha_lasso[mode_tag] = []
        for jj in range(50):
            a_ridge, vL_ridge, tL_ridge, \
            a_lasso, vL_lasso, tL_lasso, \
            vL_lr, tL_lr =  LR_analysis(dataset, 
                                        mode = reg, 
                                        do_plot = False, 
                                        n_components = 0.999)
            test_losses_LR[mode_tag].append(tL_lr)
            test_losses_ridge[mode_tag].append(tL_ridge)
            test_losses_lasso[mode_tag].append(tL_lasso)
            val_losses_LR[mode_tag].append(vL_lr)
            val_losses_ridge[mode_tag].append(vL_ridge)
            val_losses_lasso[mode_tag].append(vL_lasso)
            alpha_ridge[mode_tag].append(a_ridge)
            alpha_lasso[mode_tag].append(a_lasso)
        err_tags.append('LR ({})'.format(mode_tag))
        err_tags.append('Ridge ({})'.format(mode_tag))
        err_tags.append('Lasso ({})'.format(mode_tag))
        terr_data.append(test_losses_LR[mode_tag])
        terr_data.append(test_losses_ridge[mode_tag])
        terr_data.append(test_losses_lasso[mode_tag])
        verr_data.append(val_losses_LR[mode_tag])
        verr_data.append(val_losses_ridge[mode_tag])
        verr_data.append(val_losses_lasso[mode_tag])
        alpha_tags.append('Ridge ({})'.format(mode_tag))
        alpha_tags.append('Lasso ({})'.format(mode_tag))
        alpha_data.append(alpha_ridge[mode_tag])
        alpha_data.append(alpha_lasso[mode_tag])

terr_data = np.array(terr_data).T
verr_data = np.array(verr_data).T
alpha_data = np.array(alpha_data).T
```

```python
np.save("Plots/LinRegPlots/LossSummary.png.test_err.DATA.npy", terr_data)

np.save("Plots/LinRegPlots/LossSummary.png.val_err.DATA.npy", verr_data)

np.save("Plots/LinRegPlots/LossSummary.png.model_tags.DATA.npy", err_tags)
```
